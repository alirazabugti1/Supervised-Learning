```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("carprices.csv")
df.head()
```

```python
independent_column=df[['Mileage','Age(yrs)']]
Dependent_column=df['Sell Price($)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(independent_column,Dependent_column,test_size=0.25)
```

```python
Y_train
```

```python
reg=LinearRegression()
```

```python
reg.fit(X_train,Y_train)
```

```python
reg.predict(X_test)
```

```python
reg.score(X_test,Y_test)
```
