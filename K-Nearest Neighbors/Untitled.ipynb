```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('animals.csv')
```

```python
data
```

```python
model =KNeighborsClassifier(n_neighbors=3)
```

```python
X=data[['weight','height']]
Y=data['label']
```

```python
X
```

```python
Y
```

```python
model.fit(X,Y)
```

```python
NewAnimal=[[4,35]]
model.predict(NewAnimal)
```

```python
# Plot the data points
plt.scatter(X['weight'], X['height'], c=Y, cmap='bwr')
plt.scatter(4, 35, c='green', marker='x')
plt.xlabel("Weight (kg)")
plt.ylabel("Height (cm)")
plt.title("KNN Classification: Animal Categories")
plt.show()
```
